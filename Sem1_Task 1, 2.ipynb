### Задание 2 Сформулируйте 5-10 гипотез для выбранного продукта

**Гипотезы**

- Первая гипотеза – запуск программы лояльности с бонусами за покупки и участие в акциях может увеличить количество повторных посещений и относительно конверсии. Если внедрить предложение программы лояльности, которая будет поощрять покупки баллами, то сможем увеличить частоту повторных покупок новыми пользователями на n% и также увеличить на n% также покупку старыми пользователями.
- Вторая гипотеза - Если предложить 10% скидку на первую покупку для новых пользователей, то мы сможем повысить количество регистраций новых пользователей.
- Третья гипотеза - Если предоставить персонализированные рекомендации по продуктам новым пользователям на основе их истории просмотров и покупательского поведения, то сможем повысить показатель дохода от новых пользователей на n%.
- Четвертая гипотеза - добавление новых опций доставки или повышение скорости работы существующих способов доставки могут повысить лояльность клиентов, а также увеличить конверсию.
- Пятая гипотеза - Увелечение скорости загрузки странички товара и также рекомендации или аналогов данного товара позволит экономить время пользователей, что может позволить повысить на n% возврата пользователей к постоянным покупкам.
