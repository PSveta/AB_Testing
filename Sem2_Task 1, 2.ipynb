- Первая гипотеза – запуск программы лояльности с бонусами за покупки и участие в акциях может увеличить количество повторных посещений и относительно конверсии. - Пятая гипотеза - Увелечение скорости загрузки странички товара и также рекомендации или аналогов данного товара позволит экономить время пользователей, что может позволить повысить на n% возврата пользователей к постоянным покупкам.

### Задание 2 Составьте шаблон дизайна эксперимента для гипотезы, которая набрала больше всего баллов в практическом задании предыдущего урока

Больше всего баллов набрала Вторая гипотеза

Шаблон дизайна эксперимента:

**Гипотеза**
Если на сайте: https://www.ozon.ru/ предложить 10% скидку на первую покупку для новых пользователей, то мы сможем повысить количество регистраций новых пользователей.

**Что делаем:**

- Проведя A/B-тестирование, случайным образом разделив новых пользователей на две группы: одна группа получит код скидки 10 % на свою первую покупку, а другая группа не получит никакой скидки. 
- Через месяц измерем показатели для обеих групп и сравним результаты.
    

**На каких пользователях тестируем:** только на новых пользователях.

**Ключевые метрики для оценки эксперимента:**

1. Количество новых регистраций пользователей
2. Доход от новых пользователей
3. Стоимость предоставляемой скидки
4. Частота повторных покупок новых пользователей
5. Оценка удовлетворенности клиентов новых пользователей

**План действий в зависимости от результатов эксперимента**
• Если результат эксперимента будет положительным и прирост значений ключевых метрик будет не менее ожидаемого, то оставляем предложение на постоянной основе.
• Если значения ключевых метрик снижаются или имеют незначительный прирост (менее ожидаемого), то отменяем предложение.
